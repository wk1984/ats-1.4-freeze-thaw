```python
import sys,os
import numpy as np
from matplotlib import pyplot as plt

# This is the standard path for ATS's source directory    
try:
    import meshing_ats
except ImportError:
    sys.path.append(os.path.join(os.environ['ATS_SRC_DIR'],'tools','meshing', 'meshing_ats'))
    import meshing_ats
```

```python
# set up the surface mesh, which is a single cell
x = np.array([0.0, 1.0],'d')
z = np.array([0.0, 0.0], 'd')

# using from_Transect extrudes the x,z line in the y-direction to
# create 1 cell in y.  This results in a single cell.
m2 = meshing_ats.Mesh2D.from_Transect(x,z)

# layer extrusion
# -- data structures needed for extrusion
layer_types = []
layer_data = []
layer_ncells = []
layer_mat_ids = []
z = 0.0

layer_types.append("constant")
layer_data.append(0.10)
layer_ncells.append(10)
layer_mat_ids.append(1002)
z += 0.10

layer_types.append("constant")
layer_data.append(0.10)
layer_ncells.append(2)
layer_mat_ids.append(1002)
z += 0.10

layer_types.append("constant")
layer_data.append(0.10)
layer_ncells.append(2)
layer_mat_ids.append(1002)
z += 0.10

layer_types.append("constant")
layer_data.append(0.10)
layer_ncells.append(2)
layer_mat_ids.append(1002)
z += 0.10

layer_types.append("constant")
layer_data.append(0.10)
layer_ncells.append(2)
layer_mat_ids.append(1002)
z += 0.10

layer_types.append("constant")
layer_data.append(0.10)
layer_ncells.append(2)
layer_mat_ids.append(1002)
z += 0.10

layer_types.append("constant")
layer_data.append(0.10)
layer_ncells.append(2)
layer_mat_ids.append(1002)
z += 0.10

layer_types.append("constant")
layer_data.append(0.10)
layer_ncells.append(2)
layer_mat_ids.append(1002)
z += 0.10

layer_types.append("constant")
layer_data.append(0.10)
layer_ncells.append(2)
layer_mat_ids.append(1002)
z += 0.10

layer_types.append("constant")
layer_data.append(0.10)
layer_ncells.append(2)
layer_mat_ids.append(1002)
z += 0.10

#----------------------------------
# ^^^ TOP 100 cm ^^^
#----------------------------------

layer_types.append("constant")
layer_data.append(1.00)
layer_ncells.append(10)
layer_mat_ids.append(1002)
z += 1.0

#----------------------------------
# ^^^ TOP 200 cm ^^^
#----------------------------------

layer_types.append("constant")
layer_data.append(2.0)
layer_ncells.append(20)
layer_mat_ids.append(1002)
z += 2.0

#----------------------------------
# ^^^ TOP 400 cm ^^^
#----------------------------------

layer_types.append("constant")
layer_data.append(6.0)
layer_ncells.append(12)
layer_mat_ids.append(1002)
z += 6.0

#----------------------------------
# ^^^ TOP 10 m ^^^
#----------------------------------

layer_types.append("constant")
layer_data.append(5.0)
layer_ncells.append(5)
layer_mat_ids.append(1002)
z += 5.0

#----------------------------------
# ^^^ TOP 15 m ^^^
#----------------------------------


layer_types.append("constant")
layer_data.append(5.0)
layer_ncells.append(5)
layer_mat_ids.append(1002)
z += 5.0

#----------------------------------
# ^^^ TOP 20 m ^^^
#----------------------------------


layer_types.append("constant")
layer_data.append(10.0)
layer_ncells.append(5)
layer_mat_ids.append(1002)
z += 10.0

#----------------------------------
# ^^^ TOP 30 m ^^^
#----------------------------------


# layer_types.append("constant")
# layer_data.append(12.0)
# layer_ncells.append(4)
# layer_mat_ids.append(1002)
# z += 12.0

# print(z)

# -- mineral layer part 5 --
# keep going for 2m cells until we hit the bottom of
# the domain
layer_types.append("constant")
layer_data.append(36 - z) # depth of bottom of domain is 40 m
layer_ncells.append(int(round(layer_data[-1] / 2.0)))
layer_mat_ids.append(1002)
```

```python
meshing_ats.summarize_extrusion(layer_types, layer_data, layer_ncells, layer_mat_ids)
```

```python
# Extrude the 3D model with this structure and write to file
m3 = meshing_ats.Mesh3D.extruded_Mesh2D(m2, layer_types, 
                                        layer_data, 
                                        layer_ncells, 
                                        layer_mat_ids)
```
